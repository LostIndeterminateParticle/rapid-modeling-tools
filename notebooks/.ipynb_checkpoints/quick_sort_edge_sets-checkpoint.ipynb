```python
# Understanding quicksort
# import numpy as np

# np.random.seed(0)

# array_in = np.random.randint(0, 11, (1, 4))
# arr = [num for num in array_in[0]]
arr = [1, 3, 2, 6, 5, 0]
print(arr)

def partition(arr, low, high):
    i = (low - 1)
#     print('on partition call i: {0}'.format(i))
    pivot = arr[high]
    
    for j in range(low, high):
#         print('entered loop j: {0}'.format(j))
#         print('{0}, {1}'.format(arr[j], pivot))
        
        if arr[j] <= pivot:
#             print("arr[j] <= pivot: {0} <= {1}".format(arr[j], pivot))
#             print('this is j: {0}'.format(j))
            i += 1
#             print("i += 1: {0}".format(i))
            arr[i], arr[j] = arr[j], arr[i]
#             print("just swapped arr i and j: {0}".format(arr))
#         print(arr)
        
#     print("i: {0}".format(i))
    arr[i+1], arr[high] = arr[high], arr[i+1]
#     print("after swap in partition: {0}".format(arr))
#     print("former high, i+1: {0}, {1}".format(arr[high], arr[i+1]))

    return (i+1)

def quick_sort(arr, low, high):
#     print("quick_sort inputs: {0}, {1}, {2}".format(
#         arr, low, high))
    # This runs a check to make sure the looping is valid
    if low < high:  # if the low index less than high index, execute
        # get the partition location, returns an index
        pi = partition(arr, low, high)
        
        # in the case when low = 0 but high is -1, the fn executes
        # again
        quick_sort(arr, low, pi-1)  # sort left of the partition
        # if low > high then that fn call dies and fall back on
        # this case.
        quick_sort(arr, pi+1, high)  # sort right of partition
        
n = len(arr)
quick_sort(arr, 0, n-1)
print(arr)
# for i in range(n):
#     print(arr[i])
```

```python
# Understanding the Nuts and Bolts Problem

def compare(pivot=None, change=None):
    # so first, edge attr checked. Then look for either source or target match
    if pivot[2] == change[2]:
        if (pivot[0] == change[0]) or (pivot[1] == change[1]):
#             print(pivot, change)
            return 0  # ~equal~
#         else:
#             return 1  # ~less than~
    elif len(pivot[2]) > len(change[2]) or len(pivot[2]) == len(change[2]):
        return 1
    elif len(pivot[2]) < len(change[2]):
        return 2

def sort_nuts_and_bolts(nuts, bolts):
    
    if (len(nuts) <= 0 or len(bolts) <= 0):
        return [],[]
    
    pivot_nut = nuts[0]
    less_than_pivot_nut_pile = []
    greater_than_pivot_nut_pile = []
    bolt_equal_to_nut = None

    for bolt in bolts:
        cmp_val = compare(pivot=pivot_nut, change=bolt)
        if cmp_val == 2:
            greater_than_pivot_nut_pile.append(bolt)
        elif cmp_val == 1:
            less_than_pivot_nut_pile.append(bolt)
        else:
            bolt_equal_to_nut = bolt

    pivot_bolt = bolt_equal_to_nut
    less_than_pivot_bolt_pile = []
    greater_than_pivot_bolt_pile = []
    
    for nut in nuts:
        cmp_val = compare(pivot=pivot_bolt, change=nut)
        if cmp_val == 2:
            greater_than_pivot_bolt_pile.append(nut)
        elif cmp_val==1:
            less_than_pivot_bolt_pile.append(nut)
            
    nuts_arr = []
    bolts_arr = []
    
    lower_nuts, lower_bolts = sort_nuts_and_bolts(
        less_than_pivot_nut_pile, less_than_pivot_bolt_pile)
    
    if len(lower_nuts) > 0:
        for (n, b) in zip(lower_nuts, lower_bolts):
            nuts_arr.append(n)
            bolts_arr.append(b)
       
    nuts_arr.append(pivot_nut)
    bolts_arr.append(bolt_equal_to_nut)
    
    upper_nuts, upper_bolts = sort_nuts_and_bolts(
        greater_than_pivot_nut_pile, greater_than_pivot_bolt_pile)

    if len(upper_nuts) > 0:
        for (n, b) in zip(upper_nuts, upper_bolts):
            nuts_arr.append(n)
            bolts_arr.append(b)

    return nuts_arr, bolts_arr

base = {('s1', 't1', 'type'), ('s2', 't2', 'type'),
        ('s3', 't3', 'owner'), ('s4', 't4', 'owner'),
        ('s5', 't5', 'memberEnd'), ('s6', 't6', 'memberEnd'),}
ancestor = {('as1', 't1', 'type'), ('s2', 'at2', 'type'),
            ('as3', 't3', 'owner'), ('s4', 'at4', 'owner'),
            ('as5', 't5', 'memberEnd'),
            ('s6', 'at6', 'memberEnd'),}

base_change = base - ancestor
ance_change = ancestor - base

base_change_l = list(base_change)
ance_change_l = list(ance_change)

n, b = sort_nuts_and_bolts(base_change_l, ance_change_l)
# print('Sorted Nuts: {0}'.format(n))
# print('Sorted Bolts: {0}'.format(b))
for i in range(len(n)):
    print(n[i], b[i])
```

```python
# Understanding the Nuts and Bolts Problem

def sort_nuts_and_bolts(nuts, bolts):
    
    if (len(nuts) <= 0 or len(bolts) <= 0):
        return [],[]
    
    pivot_nut = nuts[0]
    less_than_pivot_nut_pile = []
    greater_than_pivot_nut_pile = []
    bolt_equal_to_nut = None

    for bolt in bolts:
        if bolt > pivot_nut:
            greater_than_pivot_nut_pile.append(bolt)
        elif bolt < pivot_nut:
            less_than_pivot_nut_pile.append(bolt)
        else:
            bolt_equal_to_nut = bolt

    pivot_bolt = bolt_equal_to_nut
    less_than_pivot_bolt_pile = []
    greater_than_pivot_bolt_pile = []
    
    for nut in nuts:
        if nut > pivot_bolt:
            greater_than_pivot_bolt_pile.append(nut)
        elif nut < pivot_bolt:
            less_than_pivot_bolt_pile.append(nut)
            
    nuts_arr = []
    bolts_arr = []
    
    lower_nuts, lower_bolts = sort_nuts_and_bolts(
        less_than_pivot_nut_pile, less_than_pivot_bolt_pile)
    
    if len(lower_nuts) > 0:
        for (n, b) in zip(lower_nuts, lower_bolts):
            nuts_arr.append(n)
            bolts_arr.append(b)
       
    nuts_arr.append(pivot_nut)
    bolts_arr.append(bolt_equal_to_nut)
    
    upper_nuts, upper_bolts = sort_nuts_and_bolts(
        greater_than_pivot_nut_pile, greater_than_pivot_bolt_pile)

    if len(upper_nuts) > 0:
        for (n, b) in zip(upper_nuts, upper_bolts):
            nuts_arr.append(n)
            bolts_arr.append(b)

    return nuts_arr, bolts_arr

nuts = [2, 4, 3, 0, 1]
bolts = [0, 3, 1, 2, 4]

n, b = sort_nuts_and_bolts(nuts, bolts)
print('Sorted Nuts: {0}'.format(n))
print('Sorted Bolts: {0}'.format(b))
```
